```python
#mengimport library
import matplotlib.pyplot as plt

from skimage import data
from skimage.io import imread
from skimage.color import rgb2gray
from skimage.util import invert
import numpy as np
import cv2
```

```python
#percobaan 1 : crop image

#load gambar
citraImage = cv2.imread('D:/Semester 6/Prak. Pengolahan Citra Digital/Pertemuan 9/citra.jpg')
orangeImage = cv2.imread('D:/Semester 6/Prak. Pengolahan Citra Digital/Pertemuan 9/orange.jpg')

#membuat fungsi crop pada gambar
citraCropped = citraImage.copy()
citraCropped = citraCropped[0:256,64:320]

orangeCropped = orangeImage.copy()
orangeCropped = orangeCropped[64:256,128:320]

#menampilkan gambar asli dan hasil crop dalam 1 figure
fig, axes = plt.subplots(2, 2, figsize=(12, 12))
ax = axes.ravel()
ax[0].imshow(citraImage)
ax[0].set_title("Citra Input 1")
ax[1].imshow(orangeImage, cmap='gray')
ax[1].set_title("Citra Input 2")
ax[2].imshow(citraCropped)
ax[2].set_title("Citra Output 1")
ax[3].imshow(orangeCropped, cmap='gray')
ax[3].set_title("Citra Output 2")
```

```python
#percobaan 2 : citra negative

#membalikkan gambar yang tersimpan
inv = invert(citraCropped)
print('Shape Input : ', citraCropped.shape)
print('Shape Output : ',inv.shape)

#menampilkan gambar asli, hasil invert, dan histogram dalam 1 figure
fig, axes = plt.subplots(2, 2, figsize=(12, 12))
ax = axes.ravel()

ax[0].imshow(citraCropped)
ax[0].set_title("Citra Input")

#menampilkan hasil histogram dari gambar asli
ax[1].hist(citraCropped.ravel(), bins=256)
ax[1].set_title('Histogram Input')

ax[2].imshow(inv)
ax[2].set_title('Citra Output (Inverted Image)')

#menampilkan hasil histogram dari hasil invert
ax[3].hist(inv.ravel(), bins=256)
ax[3].set_title('Histogram Output')
```

```python
#percobaan 3 : increase brightness

#mengubah tipe data menjadi float
copycitra = citraCropped.copy().astype(float)

#menambahkan parameter 100 untuk brightness pada setiap elemen aray copycitra
for i in range (copycitra.shape[0]):
    for j in range (copycitra.shape[1]):
       copycitra[i,j] += 100

copycitra[copycitra > 255] = 255
copycitra = copycitra.astype(np.uint8)

#menampilkan gambar asli, hasil brightness, dan histogram dalam 1 figure
fig, axes = plt.subplots(2, 2, figsize=(12, 12))
ax = axes.ravel()

ax[0].imshow(orangeCropped, cmap='gray')
ax[0].set_title("Citra Input")

ax[1].hist(citraCropped.ravel(), bins=256)
ax[1].set_title('Histogram Input')

ax[2].imshow(copycitra, cmap='gray')
ax[2].set_title('Citra Output (Brightnes)')

ax[3].hist(copycitra.ravel(), bins=256)
ax[3].set_title('Histogram Input')
```
