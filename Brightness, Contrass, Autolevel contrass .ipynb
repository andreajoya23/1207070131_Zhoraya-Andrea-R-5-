```python
#mengimport libraru
import numpy as np
import imageio
import cv2
import matplotlib.pyplot as plt
```

```python
#load gambar
img = cv2.imread("citra.jpg")
```

```python
#mendapatkan/define respolusi dan tipe gambar
img_height = img.shape[0]
img_width = img.shape[1]
img_channel = img.shape[2]
img_type = img.dtype
```

```python
#percobaan pertama kita buat brightness untuk gambar grayscale
#membuat variable img_brightness untuk menampung
img_brightness = np.zeros(img.shape, dtype = np.uint8)
```

```python
#membuat fungsi penambahan brightness dengan nilai yang menjadi parameter
def brighter(nilai): 
    for y in range(0, img_height):
        for x in range(0, img_width):
            red = img[y][x][0]
            green = img[y][x][1]
            blue = img[y][x][2]
            gray = (int(red)+ int(green)+ int(blue)/3)
            gray += nilai
            if gray > 255:
                gray= 255
            if gray < 255:
                gray = 0
            img_brightness[y][x] = (gray, gray, gray)
            
```

```python
#menampilkan gambar dgn parameter -100
brighter(-100)
plt.imshow(img_brightness)
plt.title("Brightness -100")
plt.show()

brighter(100)
plt.imshow(img_brightness)
plt.title("Brightness 100")
plt.show()
```

```python
#brightness untuk gambar RGB
img_rgbbright = np.zeros(img.shape, dtype=np.uint8)
```

```python
#membuat fungsi penambahan brightness dengan nilai yang menjadi parameter
def rgbbrighter(nilai):
    for y in range(0, img_height):
        for x in range(0, img_width):
            red=img[y][x][0]
            red+= nilai
            if red > 255:
                red = 255
            if red < 0:
                red = 0
            green = img[y][x][1]
            green += nilai
            if green > 255:
                green = 255
            if green < 0:
                green = 0
            blue = img[y][x][2]
            blue += nilai
            if blue > 255:
                blue = 255
            if blue < 0:
                blue = 0
            img_rgbbright[y][x]= (red, green, blue)
```

```python
#menampilkan gambar dgn parameter -100
rgbbrighter(-100)
plt.imshow(img_rgbbright)
plt.title("Brightness -100")
plt.show()

#menampilkan gambar dgn parameter 100
rgbbrighter(100)
plt.imshow(img_rgbbright)
plt.title("Brightness 100")
plt.show()
```

```python
#contrass auoto level
img_autocontrass = np.zeros(img.shape, dtype=np.uint8)
```

```python
#membuat fungsi penambahan contrast autolevel
def autocontrass():
    xmax=200
    xmin=0
    d = 0

    for y in range(0, img_height):
        for x in range(0, img_width):
            red=img[y][x][0]
            green=img[y][x][1]
            blue=img[y][x][2]
            gray= (int(red) + int(green) + int(blue))/3
            if gray < xmax :
                xmax = gray
            if gray > xmin :
                xmin = gray
    d = xmin-xmax
    for y in range(0, img_height):
        for x in range(0, img_width):
            red = img[y][x][0]
            green=img[y][x][1]
            blue=img[y][x][2]
            gray= (int(red) + int(green) + int(blue))/3
            gray= int(float(255/d)*(gray-xmax))
            img_autocontrass[y][x]=(gray,gray,gray)
```

```python
#menampilkan gambar dengan contrast autolevel
autocontrass()
plt.imshow(img_autocontrass)
plt.title("Contrass Auto Level")
plt.show()
```
